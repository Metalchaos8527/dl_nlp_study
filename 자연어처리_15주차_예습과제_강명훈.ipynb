## 날짜 형식 변환 문제를 딥러닝으로 해결하기


- 위와 같이 여러가지 형식의 날짜 표현을 규격화된 날짜 형식으로 변환하는 문제를 딥러닝으로 해결하기
- Attention 계층이 적용된 seq2seq 모델을 이용하여 해결

```python
#학습코드 구현
import numpy as np
import sys
sys.path.append(r'C:\Users\myunghoon_k\OneDrive - UOS\bitamin\dl_nlp_study\deep-learning-from-scratch-2-master')
sys.path.append(r'C:\Users\myunghoon_k\OneDrive - UOS\bitamin\dl_nlp_study\deep-learning-from-scratch-2-master\ch07')
sys.path.append(r'C:\Users\myunghoon_k\OneDrive - UOS\bitamin\dl_nlp_study\deep-learning-from-scratch-2-master\ch08')
import matplotlib.pyplot as plt
from dataset import sequence
from common.optimizer import Adam
from common.trainer import Trainer
from common.util import eval_seq2seq
from attention_seq2seq import AttentionSeq2seq
from ch07.seq2seq import Seq2seq
from ch07.peeky_seq2seq import PeekySeq2seq

#데이터 읽기
(x_train, t_train), (x_test, t_test) = sequence.load_data('date.txt')
char_to_id, id_to_char = sequence.get_vocab()

#입력 반전 (Reverse)
x_train, x_test = x_train[:, ::-1], x_test[:, ::-1]

#하이퍼 파라미터
vocab_size = len(char_to_id)
wordvec_size = 16
hidden_size = 256
batch_size = 128
max_epoch = 10
max_grad = 5.0

model = AttentionSeq2seq(vocab_size, wordvec_size, hidden_size)

optimizer = Adam()
trainer = Trainer(model, optimizer)

acc_list = []
for epoch in range(max_epoch):
    #학습
    trainer.fit(x_train, t_train, max_epoch = 1,
               batch_size = batch_size, max_grad = max_grad)
    
    correct_num = 0
    #학습 이후 테스트 진행
    for i in range(len(x_test)):
        question, correct = x_test[[i]], t_test[[i]]
        verbose = i < 10
        correct_num += eval_seq2seq(model, question, correct,
                                   id_to_char, verbose, is_reverse = True)
        
    acc = float(correct_num) / len(x_test)
    acc_list.append(acc)
    print('정확도 %.3f%%' % (acc * 100))
    
model.save_params()

#정확도 추이 그래프 표현
x = np.arange(len(acc_list))
plt.plot(x, acc_list, marker = 'o')
plt.xlabel('에폭')
plt.ylabel('정확도')
plt.ylim(-0.05, 1.05)
plt.show()
```

## 양방향 RNN (BiLSTM)
- 기존의 LSTM은 순방향 언어모델의 관점을 받아들이는 모델
- 즉 글을 왼쪽에서 오른쪽 방향으로 input하고 이를 순차적으로 학습하는 방식
- 즉 과거의 정보만을 바탕으로 미래를 보낟고 할 수 있다
    - 그런데 **미래의 정보를 가져와서 과거의 정보도 예측**한다면?
    - 즉 양방향의 정보를 모두 학습한다면 보다 좋은 분산표현 및 모델 성능을 보이지 않는가에 대한 답
    - https://stackoverflow.com/questions/43035827/whats-the-difference-between-a-bidirectional-lstm-and-an-lstm
    - 위 링크에 왜 BiLSTM이 unidirectional LSTM보다 좋을 것인지에 대한 개념적 답을 간략하게 제시함
    


- 이와 같이 BiLSTM에서는 학습 방향이 다른 2개의 LSTM계층을 이용하여 학습하고
- 두 계층의 학습 결과를 Concat하여서 **다음의 층으로 그대로 전달**하여 input데이터로 사용되게끔 한다

### ELMo (BiLM)

출처: http://www.realworldnlpbook.com/blog/improving-sentiment-analyzer-using-elmo.html

- ELMo는 Embeddings from Language Model의 약자로, **워드임베딩**을 효과적으로 하기 위해 발표된 모델이다
- 때문에 ELMo의 output은 어떤 토큰의 분산표현을 나타내게 된다, 이를 **ELMo Representaion**이라고 한다
- ELMo역시 학습 방향이 다른 2개의 LSTM, BiLSTM을 쓰지만 BiLSTM의 학습과정과 달리 학습 내에서 두개의 방향이 다른 LSTM들은 **서로의 은닉 output을 공유하지 않는다**
- 즉, 방향이 다른 LSTM들이 각자 동일한 input으로 학습을 한다음 최종 output을 **concat + weighted sum 과정**을 하여서 최종적으로 ELMo Representation을 생산하는 것이다

## ResNet

https://ganghee-lee.tistory.com/41 

- RNN을 비롯하여 신경망의 깊이가 깊어질수록 기울기 소실 & 폭발 문제가 발생
    - 모델의 일반화 성능이 떨어지는 현상
- 이를 해결하기 위한 다양한 방법론들이 등장 (새로운 모델, 기울기 클리핑 등등)


1) Residual Learning

- 네트워크 입력과 출력이 x로 동일하게 설정
- 학습하는 목적함수 $H(x)$
- 손실 $H(x) - x$가 최소화가 되도록 학습을 진행
- 그러나 여전히 기울기 소실 & 폭발 문제를 근본적으로 해결하지 못함

2) Residaul Connection

- 기존의 손실 $H(x) - x$를 $F(x)+x$가 되도록 변경
- 이를 위해 Layer마지막 단에 입력 x의 정보를 더해주는 작업을 수행, 이를 Residual Connection
- 입력 x를 더해줌으로써 기울기 역전파시 어떤 단계에서 미분을 하던 $\acute{F(x)} + 1$형태가 되므로 최소 1의 기울기를 가지게 된다
- 따라서 기울기 소실의 문제를 근본적으로 해결할 수 있게 된다
